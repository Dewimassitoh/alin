# Menggambar Bidang pada Ruang 3 Dimensi menggunakan Matplotlib

Persamaan yang akan di buat gambar tiga dimensi

\begin{align*}
x_1+2x_2+x_3=11\\  
2x_1-x_2+4x_3=4\\
3x_1+8x_2=38\\ 
\end{align*}

```python
#code
import numpy as np
import matplotlib.pyplot as plt

# Membuat Rentang x1 dan x2
rentang_x1 = np.linspace(-10,10,50) # membuat garis x1 dengan rentang -10 dampai 10, dan 50 untuk resolusi bidang yang akan kita buat
rentang_x2 = np.linspace(-10,10,50)

# Membuat grid dari rentang yang telah di buat
x2, x3 = np.meshgrid(rentang_x1, rentang_x2)

x1_1 = 1*x3+ 2*x2 - 11 
x1_2 = (4*x3 - 1*x2 - 4) 
x1_3 = (0*x3 +8*x2 - 38)/ -3 

# Membuat bidang
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

# Membuat bidang persamaan 1
# keterangan : 
# alpha = transparansi
# rstride = kepadatan gambar sumbu x1
# cstride = kepadatan gambar sumbu x2
# color = warna bidang
ax.plot_surface(x1_1,x2,x3, alpha=0.5, rstride=100, cstride=100, color="r", label="x1 + 2x2 + x3 = 11")
# membuat bidang persamaan 2
ax.plot_surface(x1_2,x2,x3, alpha=0.5, rstride=100, cstride=100, color="g", label="2x1 - x2 + 4x3 = 4")
# Membuat bidang persamaan 3
ax.plot_surface(x1_3,x2,x3, alpha=0.5, rstride=100, cstride=100, color="b", label="3x1 + 8x2  = 38")

# membuat ketrangan sumbu
ax.set_xlabel("X1")
ax.set_ylabel("X2")
ax.set_zlabel("X3")

# Menggambar bidang
plt.legend()
plt.show()
```
